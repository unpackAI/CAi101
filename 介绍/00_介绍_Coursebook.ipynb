### 什么是机器学习（ML）？

机器学习是人工智能（AI）和计算机科学的一个分支，主要是利用数据和算法来模仿人类的学习方式，逐渐提高其准确性。

学习的过程从观察或数据开始，如例子、直接经验或指令，以便在数据中寻找模式，并在未来根据我们提供的例子做出更好的决定。主要目的是让计算机在没有人类干预或协助的情况下自动学习，并相应调整行动。

这使我们能够摆脱基于规则的系统，在这种系统中，每个场景都必须预先定义并与一个行动相联系。一个独立的系统，在我们的数字时代会变得更加 "聪明"，它可以在没有明确编程的情况下做出预测或决定，这是非常强大的。

在机器学习中，有两种基本方法让机器从数据中 "学习"。这两种方法被称为**监督学习**和**无监督学习**。

**监督学习**。监督学习是一种机器学习方法，它的定义是使用标记的数据集。这意味着，对于每一个数据输入，我们也提供结果/输出，以便程序将其预测与实际结果进行比较，从而调整自己。使用标记的输入和输出，该模型可以衡量其准确性，并随着时间的推移而学习。

>*举例说明：*一个程序接收一件衣服的图像作为输入，并被期望预测它是什么类型的衣服（如T恤、鞋子等）。它在一个有标签的服装图像数据集上进行训练，告诉模型哪些照片是T恤衫、鞋子、夹克衫等。

**无监督学习**。无监督学习使用机器学习算法来分析和聚类*未标记的*数据集。这些算法发现数据中的隐藏模式，而不需要人类的干预（因此，它们是 "无监督 "的）。

> *举例说明：*一个程序收到一个没有标记的服装图像数据集，并期望在图像中找到模式。由于是无监督的，它可以在各种图像中找到最终代表鞋子、夹克等的模式。

在接下来的几周里，这两者之间的区别将变得更加清晰，届时我们将对有监督和无监督的机器学习应用进行研究。

### 什么是深度学习（DL）？

深度学习是机器学习的一个子集，它利用称为**神经网络**的计算系统，从数据中学习和寻找模式，以高精确度执行任务。


#### 什么是神经网络？

神经网络是一系列处理数据的算法，能够灵活地调整自己以达到最高的准确性。它们的名称和结构受到人脑的启发，模仿了生物神经元相互之间的信号方式。

在神经网络中，"神经元 "是一个节点，一个能够处理数据的计算单元。每个节点都有一个相关的权重，这个参数可以根据传入的数据灵活地改变，这也将影响整个网络的结果。整个网络进行自我优化以达到更好的效果。

一个神经由至少3层组成。

1. **输入层**将初始数据带入系统，由后续层进一步处理。
2. **隐藏层**位于输入层和输出层之间，对输入施加权重，并将其引向输出。
3. **输出层**负责产生最终结果。输出层接收从前面一层传来的输入，并从一系列的输入中计算出一个决定。

数据在神经网络中被多次传递。

#### 什么是GPU？

一个**图形处理单元（GPU）**，也被称为图形卡，是一个处理单元，在运行神经网络和训练模型时被使用。

它是计算机中一种特殊的处理器，可以同时处理数以千计的单一任务，特别是为在计算机上显示3D环境玩游戏而设计。这些相同的基本任务与神经网络的工作非常相似，这样，GPU运行神经网络的速度比普通CPU快数百倍。所有现代计算机都包含一个GPU，但很少有计算机包含深度学习所需的正确类型的GPU。

让我们最后来训练我们的模型。如果你观察下面的代码，你将能够看到`dls`，`resnet34`和`error_rate`。

`dls`： 我们将所有处理过的数据打包到我们的 "DataLoaders"（将数据加载到我们的模型中），我们使用快捷键`dls`。

`resnet34`：Resnet是一种预先构建的神经网络，我们可以轻松使用。"34"代表该神经网络的总层数。

`误差率`：误差率是我们定义的指标，用于观察我们模型的性能。

```python
learn = cnn_learner(dls, resnet34, metrics=error_rate)
learn.fine_tune(1)
```

哇! 我们已经实现了0.6%的`错误率'。

但是，我们怎么知道这个模型是否好呢？错误率的计算，是指被错误识别的图像比例。错误率作为我们的衡量标准--我们对模型质量的衡量，被选择为直观和可理解的。正如你所看到的，这个模型几乎是完美的，尽管训练时间非常短。事实上，你所取得的准确度已经远远超过了10年前任何人所取得的准确度!

### 我的模型做了什么？

```python
#@title
gv('''ordering=in
model[shape=box3d width=1 height=0.7 label="Neural Network + Parameters"]
inputs->model->predictions; labels->loss; predictions->loss
loss->model[constraint=false label=update]''')
```

上面的图片是一个概念化的模型，它经历了什么样的过程，以达到如此低的 "错误率"，并成功预测输入的图片是猫还是狗。让我们一起回顾一下这个过程。

1. 输入: 输入是我们下载的数据集的数据。就像上面所说的，我们必须在将数据 "输入 "到我们的神经网络之前准备好这些数据。
2. 神经网络。就像我们上面所学到的，数据被传入神经网络的第一个输入层，并经过每一层。
3. 预测: 根据输出层的输出，产生一个预测。
4. 标签: 除了预测之外，该模型还查看图像的实际内容，这些内容是通过标签描述的。
5. 损失: 现在我们有了一个预测和一个标签，我们对它们进行比较并计算出损失。预测的错误越多，损失就越大。预测越错，损失就越大。
6. 更新: 根据损失（如果高或低），模型更新其参数以达到更低的损失（和更高的准确性）。
7. 参数: 影响预测的数值，随着数据在模型中的每一次迭代而得到改善。

### 机器学习固有的局限性

从这幅图中，我们现在可以看到关于训练深度学习模型的一些基本情况。

- 没有数据就无法创建一个模型。
- 一个模型只能在用于训练它的输入数据中学习操作模式。
- 这种学习方法的**预测**只能作为参考，但并非百分百准确。
- 仅仅有输入数据的例子是不够的；我们还需要这些数据的*标签（例如，狗和猫的图片不足以训练一个模型；我们需要为每张图片贴上标签，说明哪些是狗，哪些是猫）。

一般来说，我们看到大多数组织说他们没有足够的数据，实际上是指他们没有足够的*标签的数据。如果任何组织有兴趣在实践中用一个模型做一些事情，那么他们大概有一些输入，他们计划针对这些模型来运行。而且，他们可能已经用其他方式做了一段时间（例如，手动，或用一些启发式程序），所以他们有来自这些过程的数据。例如，一个放射科诊所几乎肯定会有一个医疗扫描档案（因为他们需要能够检查他们的病人在一段时间内的进展情况），但这些扫描可能没有包含诊断或干预措施清单的结构化标签（因为放射科医生通常创建自由文本自然语言报告，而不是结构化数据）。我们将在本书中经常讨论标签方法，因为它是实践中的一个重要问题。

由于这类机器学习模型只能进行*预测*（即试图复制标签），这可能导致组织目标和模型能力之间的巨大差距。例如，在本书中，你将学习如何创建一个*推荐系统*，可以预测用户可能购买的产品。这通常用于电子商务中，例如通过显示排名最高的商品来定制显示在主页上的产品。但这样的模型通常是通过查看用户和他们的购买历史（*输入*）以及他们继续购买或查看的东西（*标签*）来创建的，这意味着该模型可能会告诉你用户已经拥有或已经知道的产品，而不是他们最有可能感兴趣的新产品。这与比如说你当地书商的专家可能做的非常不同，他们会问一些问题来了解你的品味，然后告诉你一些你以前从未听说过的作者或系列。

另一个关键的见解来自于考虑一个模型如何与它的环境互动。这可以创造*反馈循环*，如这里所描述的。

- 一个*预测性警务*模型是根据过去的逮捕地点创建的。在实践中，这实际上不是在预测犯罪，而是在预测逮捕，因此部分地只是反映了现有警务过程中的偏见。
- 然后，执法人员可能会使用该模型来决定将他们的警察活动集中在哪里，导致这些地区的逮捕人数增加。
- 关于这些额外逮捕的数据将被反馈到模型的未来版本中。

这是一个*正反馈循环*，模型用得越多，数据就越有偏向，使模型更加偏向，以此类推。

反馈环路在商业环境中也会产生问题。例如，一个视频推荐系统可能会偏向于推荐那些最大的视频观看者所消费的内容（例如，阴谋论者和极端主义者往往比一般人观看更多的在线视频内容），导致这些用户增加他们的视频消费，从而导致更多的这类视频被推荐。

在接下来的几周里，我们将把机器学习工作流程应用于我们所有的ML应用。它是任何机器学习模型的组成部分。

1. *将商业目标转化为机器学习任务：*定义你的目标、数据集、你打算建立的模型和你想衡量的指标。重要的是要做研究，看看是否已经存在标记的数据，它来自哪里，以确定你应该如何为你的模型定义它。你的输入和输出是什么？

2. *收集和构建你的数据集：* 这是机器学习工程师的工作中最耗时的工作。在定义了你的数据集之后，你必须开始（1）收集数据，（2）确定特征和标签来源，（3）选择抽样策略，以及（4）将数据分成不同的集合。

3. *数据转换：*数据转换对于数据兼容是绝对必要的。收集和构建的数据必须进行转换，以使模型能够处理它。除了转换数据类型，将输入的大小调整为固定的大小，以及其他转换，你还必须探索和清理你的数据（即检查数据行，修复条目等）。

4. *训练你的模型：*训练意味着创建或学习模型。也就是说，你向模型展示标记的例子，并使模型逐渐学习特征和标签之间的关系。你会意识到，这个过程是自动的，你必须密切关注你在第1步中定义的关键指标。

5. *解释模型并进行预测：*一旦模型被成功训练，你将能够使用各种可视化和解释方法来解释其性能。机器学习推理是指将训练好的模型应用于它以前从未见过的未标记的例子，并对数据进行实际预测。
