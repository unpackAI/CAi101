Âú®‰∏ãËΩΩ‰∫Ü‰∏Ä‰∏™ÂÖÖÊª°ÂõæÂÉèÁöÑÊï∞ÊçÆÈõÜÂêéÔºåÊàë‰ª¨Áé∞Âú®ÂøÖÈ°ª‰ª•‰∏ÄÁßçÂèØ‰ª• "ÂñÇ "ÁªôÊàë‰ª¨ÁöÑÊ®°ÂûãÁöÑÊñπÂºèÊù•ÂáÜÂ§áÊï∞ÊçÆ„ÄÇËØ∑ËøêË°å‰∏ãÈù¢ÁöÑÂçïÂÖÉÊ†º!

```python
#@title Prepare My Data for my model
def is_cat(x): return x[0].isupper()
dls = ImageDataLoaders.from_name_func(
    path, get_image_files(path), valid_pct=0.2, seed=42,
    label_func=is_cat, item_tfms=Resize(224))

```

ËÆ©Êàë‰ª¨ÊúÄÂêéÊù•ËÆ≠ÁªÉÊàë‰ª¨ÁöÑÊ®°Âûã„ÄÇÂ¶ÇÊûú‰Ω†ËßÇÂØü‰∏ãÈù¢ÁöÑ‰ª£Á†ÅÔºå‰Ω†Â∞ÜËÉΩÂ§üÁúãÂà∞`dls`Ôºå`resnet34`Âíå`error_rate`„ÄÇ

`dls`Ôºö Êàë‰ª¨Â∞ÜÊâÄÊúâÂ§ÑÁêÜËøáÁöÑÊï∞ÊçÆÊâìÂåÖÂà∞Êàë‰ª¨ÁöÑ "DataLoaders"ÔºàÂ∞ÜÊï∞ÊçÆÂä†ËΩΩÂà∞Êàë‰ª¨ÁöÑÊ®°Âûã‰∏≠ÔºâÔºåÊàë‰ª¨‰ΩøÁî®Âø´Êç∑ÈîÆ`dls`„ÄÇ

`resnet34`ÔºöResnetÊòØ‰∏ÄÁßçÈ¢ÑÂÖàÊûÑÂª∫ÁöÑÁ•ûÁªèÁΩëÁªúÔºåÊàë‰ª¨ÂèØ‰ª•ËΩªÊùæ‰ΩøÁî®„ÄÇ"34"‰ª£Ë°®ËØ•Á•ûÁªèÁΩëÁªúÁöÑÊÄªÂ±ÇÊï∞„ÄÇ

`ËØØÂ∑ÆÁéá`ÔºöËØØÂ∑ÆÁéáÊòØÊàë‰ª¨ÂÆö‰πâÁöÑÊåáÊ†áÔºåÁî®‰∫éËßÇÂØüÊàë‰ª¨Ê®°ÂûãÁöÑÊÄßËÉΩ„ÄÇ

```python
learn = cnn_learner(dls, resnet34, metrics=error_rate)
learn.fine_tune(1)
```

Âìá! Êàë‰ª¨Â∑≤ÁªèÂÆûÁé∞‰∫Ü0.6%ÁöÑ`ÈîôËØØÁéá'„ÄÇ

‰ΩÜÊòØÔºåÊàë‰ª¨ÊÄé‰πàÁü•ÈÅìËøô‰∏™Ê®°ÂûãÊòØÂê¶Â•ΩÂë¢ÔºüÈîôËØØÁéáÁöÑËÆ°ÁÆóÔºåÊòØÊåáË¢´ÈîôËØØËØÜÂà´ÁöÑÂõæÂÉèÊØî‰æã„ÄÇÈîôËØØÁéá‰Ωú‰∏∫Êàë‰ª¨ÁöÑË°°ÈáèÊ†áÂáÜ--Êàë‰ª¨ÂØπÊ®°ÂûãË¥®ÈáèÁöÑË°°ÈáèÔºåË¢´ÈÄâÊã©‰∏∫Áõ¥ËßÇÂíåÂèØÁêÜËß£ÁöÑ„ÄÇÊ≠£Â¶Ç‰Ω†ÊâÄÁúãÂà∞ÁöÑÔºåËøô‰∏™Ê®°ÂûãÂá†‰πéÊòØÂÆåÁæéÁöÑÔºåÂ∞ΩÁÆ°ËÆ≠ÁªÉÊó∂Èó¥ÈùûÂ∏∏Áü≠„ÄÇ‰∫ãÂÆû‰∏äÔºå‰Ω†ÊâÄÂèñÂæóÁöÑÂáÜÁ°ÆÂ∫¶Â∑≤ÁªèËøúËøúË∂ÖËøá‰∫Ü10Âπ¥Ââç‰ªª‰Ωï‰∫∫ÊâÄÂèñÂæóÁöÑÂáÜÁ°ÆÂ∫¶!

ÊúÄÂêéÔºåËÆ©Êàë‰ª¨Ê£ÄÊü•‰∏Ä‰∏ãËøô‰∏™Ê®°ÂûãÊòØÂê¶ÁúüÁöÑÊúâÊïà„ÄÇÂéªÊâæ‰∏ÄÂº†ÁãóÊàñÁå´ÁöÑÁÖßÁâáÔºõÂ¶ÇÊûú‰Ω†ÊâãÂ§¥Ê≤°ÊúâÔºåÂ∞±Âú®Ë∞∑Ê≠åÂõæÁâá‰∏äÊêúÁ¥¢Âπ∂‰∏ãËΩΩ‰∏ÄÂº†‰Ω†Âú®ÈÇ£ÈáåÊâæÂà∞ÁöÑÂõæÁâá„ÄÇ

Áé∞Âú®ÊâßË°åÂÆö‰πâ‰∫Ü`uploader`ÁöÑÂçïÂÖÉ„ÄÇÂÆÉÂ∞ÜËæìÂá∫‰∏Ä‰∏™ÊåâÈíÆÔºöÁÇπÂáªÂÆÉÔºåÈÄâÊã©‰Ω†ÊÉ≥ÂàÜÁ±ªÁöÑÂõæÁâáÔºåÂÆÉÂ∞±‰ºöË¢´‰∏ä‰º†„ÄÇ

```python
# Upload an image from your computer
uploader = widgets.FileUpload()
uploader
```

Â¶ÇÊûú‰Ω†Âú®‰∏äÈù¢Ê≠£Á°ÆÂú∞‰∏ä‰º†‰∫ÜÂõæÁâáÔºå‰Ω†Â∫îËØ•ËÉΩÂú®‰∏ãÈù¢ÁúãÂà∞ÂÆÉ„ÄÇ

```python
img = PILImage.create(uploader.data[0])
img
```

```python
is_cat, _, probs = learn.predict(img)
print(f"ËøôÊòØÁå´Âêó?: {is_cat}.")
print(f"ËøôÊòØ‰∏ÄÂè™Áå´ÁöÑÊ¶ÇÁéá: {probs[1].item():.6f}")
```
